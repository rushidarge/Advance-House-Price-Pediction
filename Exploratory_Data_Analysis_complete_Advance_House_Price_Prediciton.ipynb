# Correlation

```
numerical_features = df.select_dtypes(include=[np.number])
numerical_features.columns
```

```
categorical_features = df.select_dtypes(include=[np.object])
categorical_features.columns
```

```
correlation = numerical_features.corr()
print(correlation["SalePrice"].sort_values(ascending=False))
```

```
less_corr = ['LotFrontage', 'WoodDeckSF', '2ndFlrSF', 'OpenPorchSF', 'HalfBath',
       'LotArea', 'BsmtFullBath', 'BsmtUnfSF', 'BedroomAbvGr', 'ScreenPorch',
       'PoolArea', 'MoSold', '3SsnPorch', 'BsmtFinSF2', 'BsmtHalfBath',
       'MiscVal', 'Id', 'LowQualFinSF', 'YrSold', 'OverallCond', 'MSSubClass',
       'EnclosedPorch', 'KitchenAbvGr']
```

```
f, ax = plt.subplots(figsize=(12,10))
plt.title("Correlation of numeric features with sale price",y=1,size=16)
sns.heatmap(correlation,square=True,vmax=0.8)
```

```
cols = correlation.nlargest(5, "SalePrice")["SalePrice"].index
```

```
np.corrcoef(df[cols].T)
```

```
np.corrcoef(df[cols].values.T)
```

```
k = 11
cols = correlation.nlargest(k, "SalePrice")["SalePrice"].index
print(cols)
cm = np.corrcoef(df[cols].T)
f, ax = plt.subplots(figsize = (10,12))
sns.heatmap(cm, annot=True, cmap="viridis", square=True, linecolor="white", linewidths=0.01, xticklabels= cols.values, yticklabels=cols.values)
```

# Outliers

```
df["SalePrice"].describe()
```

```
sns.boxplot(x=df["SalePrice"])
```

```
df.shape
```

```
first_quantile = df["SalePrice"].quantile(.25)
third_quantile = df["SalePrice"].quantile(.75)
IQR = third_quantile - first_quantile
```

```
new_boundry = third_quantile + 3*IQR
```

```
df.drop(df[df["SalePrice"]>new_boundry].index,axis=0,inplace=True)
```

```
df.shape
```

# Remove bad features

```
df.drop(less_corr,axis=1,inplace=True)
```

```
df.shape
```

```
more_null = df.isnull().mean().sort_values(ascending=False).index[0:5]
df.isnull().mean().sort_values(ascending=False).index[0:5]
```

```
df.drop(more_null,axis=1,inplace=True)
```
